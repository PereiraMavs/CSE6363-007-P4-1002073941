```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
```

```python
# Function to read data from the file

with open('P3input2024_pre.txt', 'r') as f:
    data = f.read().splitlines()

# Extract features
x = [[float(line.split()[0]), float(line.split()[1])] for line in data]
    
# Extract labels
y = [line.split()[2] for line in data]
y = np.array([1 if i == '0' else 0 for i in y])

# Example usage:
# Replace 'P3input2024_pre.txt' with the actual path to your data file


# Display a preview of the data
print("Features (x):", x[:5])  # Show the first 5 feature pairs
print("Labels (y):", y[:5])    # Show the first 5 labels
```

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
x_min, x_max = 0, 19
y_min, y_max = 0, 19
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
```

```python
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
x=np.array(x)
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.title("Logistic Regression Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
